## Combine LHD and uncertainty sampling training sets

````python
filename1 = '../../data/TEOBv4QM_20hz/h_train_corners_lhd.hdf5'
filename2 = '../../data/TEOBv4QM_20hz/h_train_uncsamp.hdf5'
filename_join = '../../data/TEOBv4QM_20hz/h_train_lhd_uncsamp.hdf5'
ws.join_waveform_sets(filename1, filename2, filename_join)

filename1 = '../../data/TEOBv4QM_20hz/dh_train_corners_lhd.hdf5'
filename2 = '../../data/TEOBv4QM_20hz/dh_train_uncsamp.hdf5'
filename_join = '../../data/TEOBv4QM_20hz/dh_train_lhd_uncsamp.hdf5'
ws.join_waveform_sets(filename1, filename2, filename_join)
````

````python
h_train = ws.HDF5WaveformSet('../../data/TEOBv4QM_20hz/h_train_lhd_uncsamp.hdf5')
dh_train = ws.HDF5WaveformSet('../../data/TEOBv4QM_20hz/dh_train_lhd_uncsamp.hdf5')
````

````python
fig, (ax1, ax2) = wave.plot_waveforms_fd([h_train[i] for i in range(len(h_train))])
ax1.set_yscale('log')
ax1.set_ylabel(r'$A$')
ax2.set_ylabel(r'$\Phi$')
ax2.set_xlabel(r'$Mf$')
ax1.grid()
ax2.grid()

fig, (ax1, ax2) = wave.plot_waveforms_fd([dh_train[i] for i in range(len(dh_train))])
ax1.set_ylabel(r'$\ln(A/A_{\rm F2})$')
ax2.set_ylabel(r'$\Phi-\Phi_{\rm F2}$')
ax2.set_xlabel(r'$Mf$')
ax1.grid()
ax2.grid()
````
